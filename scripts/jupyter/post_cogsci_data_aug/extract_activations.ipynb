## compute mean activation for category

```python
dir = '../../../data/alexnet_hp_tuned_all_concepts/model_activations_pre_final_all_test_images/'
# for all csv files in dir, get the df, and then get the mean of all the activations for each concept in the df. The concept is the index of the df
# do this for all the csv files in the dir
# save the results to a csv file in ../../../data/alexnet_hp_tuned_all_concepts/pre_final_mean_concept_activations_test/

import pandas as pd
import numpy as np
import os
import pickle
import sys
sys.path.append('/home/siddsuresh97/Projects/vision_robustness_using_semantic_norms')
from src.utils import cogsci_helper_functions

dir = '../../../data/alexnet_hp_tuned_all_concepts/model_activations_pre_final_all_test_images/'
if not os.path.exists('../../../data/alexnet_hp_tuned_all_concepts/pre_final_mean_concept_activations_test/'):
    os.makedirs('../../../data/alexnet_hp_tuned_all_concepts/pre_final_mean_concept_activations_test/')
for file in os.listdir(dir):
    print(file)
    df = pd.read_csv(dir+file)
    # take the mean of all activations in 'Unnamed: 0' columns
    import ipdb;ipdb.set_trace()
    df = df.groupby('Unnamed: 0').apply(lambda x: x.select_dtypes(include=[np.number]).mean())
    df.to_csv('../../../data/alexnet_hp_tuned_all_concepts/pre_final_mean_concept_activations_test/{file}'.format(file=file))
```
